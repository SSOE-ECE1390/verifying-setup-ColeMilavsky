```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Load grayscale image
image_path = Path("Data/Apollo_11_Launch.jpg")
image = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)
```

```python
# Display Image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Size of image
print("Image shape:", image.shape)

# Crop Image and display
cropped_image = image[100:620, 100:1180]
plt.imshow(cropped_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Increase brightness of cropped image
# alpha effects contrast (multiplies), beta effects brightness (adds)
brightened_image = cv2.convertScaleAbs(cropped_image, alpha=1, beta=100)

# Display brightened image
plt.imshow(brightened_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Rotate image 90 degrees
rotated_image = cv2.rotate(brightened_image, cv2.ROTATE_90_CLOCKWISE)

# Display rotated image
plt.imshow(rotated_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Save modified image
output_path = Path("Data/Apollo_11_Launch_Modified.jpg")
cv2.imwrite(str(output_path), rotated_image)
```
